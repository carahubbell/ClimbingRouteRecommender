```python
# Get potential imports

from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_colwidth', 500)
import time
import requests
import random
```

```python
# From MP's Partner Finder, locate 500 users within 25 miles of Golden (80401) who will be our sample community.

page = requests.get("https://www.mountainproject.com/partner-finder/results?min-age=-&max-age=-&location=80401&distance=25&_token=FpjSuc4yZku93ZQgno7kqHn32OS0cAjyYsDmkWeU")
page
```

```python
# Collect the HTML code

soup = bs(page.content)
```

```python
# Select only the instances of class = 'name-location' which holds all user details

urls = soup.find_all(class_='name-location')
urls
```

```python
# Take HTML code as a string and parse on / so we can take subsets of each user's URL

users = str(urls).split("/")
users
```

```python
# Since the user IDs are located in position 4, 13, 22, 31, etc, we take each value that has a remainder of 4 when divided by 9.

def GetIDs(users = users, n = 4):
    ID_list = []
    for i in range(len(users)):
        if i % 9 == n:
            ID_list.append(users[i])
    return ID_list

user_IDs = GetIDs(users, n = 4)
len(user_IDs) # confirm we have 500 climbers
```

```python
# View the user IDs
user_IDs
```

```python
# Since the user names are located in position 5, 14, 23, 32, etc, we take each value that has a remainder of 5 when divided by 9.

def GetNames(users = users, n = 5):
    ID_list = []
    for i in range(len(users)):
        if i % 9 == n:
            ID_list.append(users[i])
    return ID_list

user_names = GetNames(users, n = 5)
len(user_names) # confirm we have 500 climbers
```

```python
# View the users' names
user_names = [i.split('">', 1)[0] for i in user_names]
user_names
```

```python
# Build DF of ID and name

userdf = pd.DataFrame({'user_ID':user_IDs,
                      'username':user_names})
userdf
```

```python
# Export the dataframe

userdf.to_csv(r'"C:\Users\hubbell\Desktop\Data Science\Climbing Recommender\climbers_dataframe.csv', index = False, header = True)
```

```python
# Function to generate all user URLs to be used to export ticks

def url_builder(userdf):
    ticks = pd.DataFrame()
    for i in range(len(userdf)):
        url = 'https://www.mountainproject.com/user/' + str(userdf['user_ID'].iloc[i]) + '/' + str(userdf['username'].iloc[i]) + '/tick-export'
        data = pd.read_csv(url)
        data['user_ID'] = userdf['user_ID'].iloc[i]
        ticks = ticks.append(data)
    return ticks

ticks = url_builder(userdf)
ticks = ticks[['user_ID', 'Route', 'Rating', 'URL', 'Avg Stars', 'Your Stars']]
```

```python
# Export the dataframe

ticks.to_csv(r"C:\Users\hubbell\Desktop\Data Science\Climbing Recommender\tick_dataframe.csv", index = False, header = True)
```
